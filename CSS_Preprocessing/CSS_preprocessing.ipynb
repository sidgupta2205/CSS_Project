```python
import os
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python

def load_dataset( dir_path ):
  img_list = []
  for filename in os.listdir(dir_path):
    if(filename.endswith("jpg") or filename.endswith("png") or filename.endswith("jpeg")):
      # print( filename )
      # display_image( dir_path, filename )
      img_list.append(filename)
  return img_list    
```

```python
def read_image( file_name ):
  # print( file_name )
  # file_name = os.path.join(dir_path,file_name) 
  img = cv2.imread(file_name)
  return img

def display_image( img ):
  cv2_imshow(img)

def img_resizing(img):
  # scale_percent = 60 # percent of original size
  # width = int(img.shape[1] * scale_percent / 100)
  # height = int(img.shape[0] * scale_percent / 100)
  width = 200
  height = 200
  dim = (width, height) 
  resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
  return resized


def gauss_blur_images( img ):
  # Gaussian Blurring
  # Again, you can change the kernel size
  gausBlur = cv2.GaussianBlur(img, (5,5),0)
  return gausBlur

def median_blur_images( img ):
  medBlur = cv2.medianBlur( img, 5 )
  return medBlur

def bilateral_filter_images( img ):
  bilFilter = cv2.bilateralFilter(img,9,75,75)
  return bilFilter



def simple_thresholding( img ):
  #ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
  #ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
  #ret, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC)
  ret, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO)
  #ret, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)
  return thresh4

def normalization( img ):
  norm_img = np.zeros((534,800, 3))
  final_img = cv2.normalize(img,  norm_img, 0, 255, cv2.NORM_MINMAX)
  return final_img
```

```python
dir_path = "Images"
#load_dataset( dir_path )
sample_image_name = "103.jpg"
sample_image = read_image( sample_image_name )
#display_image( sample_image )
resized_sample_image = img_resizing( sample_image )
display_image( resized_sample_image )


print( "Normalize---------------------------------------------" )
normalized_img = normalization( resized_sample_image )
display_image( normalized_img )

'''
print( "gaus_blurred_image------------------------------------------" )
gaus_blurred_image = gauss_blur_images( resized_sample_image )
display_image( gaus_blurred_image )

print( "med_blurred_image------------------------------------------" )
med_blurred_image = median_blur_images( resized_sample_image )
display_image( med_blurred_image )
'''

print( "simple_thresh_image------------------------------------------" )
thresh4 = simple_thresholding( normalized_img )
print( "thresh4---------------------------------------------" )
display_image( thresh4 )


print( "bil_filter_image------------------------------------------" )
bil_filter_image = bilateral_filter_images( normalized_img )
display_image( bil_filter_image )


print("img shape", sample_image.shape)
```
